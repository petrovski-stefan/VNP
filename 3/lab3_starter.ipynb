```python
df.drop(index= df[df['FastCharge_KmH']=="-"].index,inplace=True)
```

```python
df[df['FastCharge_KmH']=="-"]
```

```python
types = df.dtypes
cols = df.columns
cols = [col for col in cols if types[col]=="object"]
# RapidCharge, PowerTrain, PlugType, BodyStyle, Segment
print(df[cols].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['PlugType'] = label_encoder.fit_transform(df['PlugType'])
df['RapidCharge'] = label_encoder.fit_transform(df['RapidCharge'])
df['PowerTrain'] = label_encoder.fit_transform(df['PowerTrain'])
df['BodyStyle'] = label_encoder.fit_transform(df['BodyStyle'])
df['Segment'] = label_encoder.fit_transform(df['Segment'])

df[cols]
```

```python
from sklearn.model_selection import train_test_split

df['Brand'] = label_encoder.fit_transform(df['Brand'])

df.drop(columns=["Model"],inplace=True)

X = df.drop(columns=['PriceEuro'])
Y = df['PriceEuro']

```

```python
x_train, x_test, y_train,y_test = train_test_split(X,Y,test_size=0.2)


x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
dt_regressor = DecisionTreeRegressor(max_depth=3, min_samples_split=10,max_features="log2")


dt_regressor.fit(x_train,y_train)

y_pred = dt_regressor.predict(x_test)

print(f"R squared: {r2_score(y_test,y_pred)}")
```

```python
from xgboost import XGBRegressor

xgb_regressor = XGBRegressor(n_estimators=1000,max_depth=5,objective="reg:squarederror")
df['FastCharge_KmH'] = df['FastCharge_KmH'].astype(float)
categorical_cols = ['RapidCharge', 'PowerTrain', 'PlugType', 'BodyStyle', 'Segment',"Brand"]
df[categorical_cols] = df[categorical_cols].astype('category')

xgb_regressor.fit(x_train,y_train)

y_pred = xgb_regressor.predict(x_test)
print(r2_score(y_test,y_pred))


```
